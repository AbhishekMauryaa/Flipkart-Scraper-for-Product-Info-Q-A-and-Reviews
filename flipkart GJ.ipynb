```python
# Define the function to generate Flipkart URLs from PIDs with an additional parameter
def generate_flipkart_url(pid):
    base_url = 'https://www.flipkart.com/product/p/itme?pid='
    additional_param = '&lid=abcd'  # Define the additional query parameter
    return base_url + pid + additional_param  # Append the PID and the additional parameter to the base URL


# Generate product URLs with the additional query parameter
product_urls = [generate_flipkart_url(pid) for pid in pids]

# Print the generated product URLs with the additional query parameter
for url in product_urls:
    print(url)
```

```python
from bs4 import BeautifulSoup
import time
from tqdm import tqdm  # Make sure to import or define tqdm and selenium webdriver correctly

# Assuming driver setup and product_urls definition are done elsewhere
# e.g., from selenium import webdriver
# driver = webdriver.Chrome()

# Initialize lists to store data
pids = []
product_titles = []
landing_image = []
category_node = []
fulfillment_mode = []
prices=[]
MRPs=[]
offers = []
warranty = []
variations = []
highlights = []
payment_modes = []
description_list = []
A_plus_contents=[]
product_specifications = []
ratings_list = []
count_of_ratings = []
count_of_reviews = []
all_ratings_by_star = []
reviews_urlss = []
QnAs_urlss = []
placeholder_images = []
stocks=[]


# Loop through each URL
for url in tqdm(product_urls):
    driver.get(url)
    time.sleep(1)  # Be mindful of the website's load time and terms of use
    html_data = BeautifulSoup(driver.page_source, 'html.parser')
    product = html_data.find('div', {'class': '_1YokD2 _2GoDe3'})
    parsed_url = urlparse(url)
    query_params = parse_qs(parsed_url.query)
    pid = query_params.get('pid', ['N/A'])[0]  # Default to 'N/A' if not found
    pids.append(pid)
    
    if product:


        
## Extract product title
        product_title = product.find('span', class_='B_NuCI').text.strip()
        product_titles.append(product_title)


##landing image
        # Find all img elements with the specified class attributes
        img_elements = product.find_all('img', class_='_396cs4 _2amPTt _3qGmMb')
        
        # Iterate over each img element found
        for img_element in img_elements:
            # Extract the src attribute
            image_url = img_element.get('src')
            if image_url:  # Check if src attribute is present
                landing_image.append(image_url)
        
        # If landing_image is empty, use 'na' as a placeholder
        if not landing_image:
            landing_image.append('na')
        
## Extract category nodes
        category_node_divs = product.find_all('div', class_='_3GIHBu')
        
        # Extract text content from each category node
        if category_node_divs:
            # Extracted category names as a list of strings
            categories = [category_node_div.text.strip() for category_node_div in category_node_divs]
            
            # Concatenate the category names with ">"
            formatted_category = ' > '.join(categories)
            
            # Append the formatted category to the category_node list
            category_node.append(formatted_category)
        else:
            category_node.append('N/A')  # Append 'N/A' if no category name is found


## Extract Stock
        stock_element = product.find('div', class_='_16FRp0')
        if stock_element:
            stock = stock_element.text.strip()
            stocks.append(stock)
        else:
            # If the element is not found, append 'N/A' or another placeholder to indicate the absence of stock information
            stocks.append('In Stock')

## Extract product price
        ## Extract product price
        try:
            # Find the div element with the specified class
            price_div = product.find('div', class_='_30jeq3 _16Jk6d')
        
            # Check if the div element is found and extract its text
            if price_div:
                prices.append(price_div.text.strip())
            else:
                # If price_div is not found, append 'N/A'
                prices.append('N/A')
        except Exception as e:
            # If an exception occurs (e.g., element not found or unexpected HTML structure), handle it gracefully
            print(f"Error occurred while extracting price: {str(e)}")
            prices.append('N/A')


## Extract mrp price
        # Find the div element with the specified class
        MRP_div = product.find('div', class_='_3I9_wc _2p6lqe')

        # Check if the div element is found and extract its text
        if MRP_div:
            MRPs.append(MRP_div.text.strip())
        else:
            MRPs.append('N/A')

##Fulfillment mode
        
        image_element = product.find('img', class_='jMnjzX')
        if image_element:
            fulfillment_mode.append('Flipkart Assured')
        else:
            fulfillment_mode.append('N/A')


            
##Offers
        # Find all the li elements with the specified class
        offer_li_elements = product.find_all('li', class_='_16eBzU col')

        # Extract offer text from each li element and concatenate into a single string
        offer_texts = []
        for offer_li in offer_li_elements:
            offer_spans = offer_li.find_all('span')
            offer_texts.extend([span.text.strip() for span in offer_spans])
        
        # Concatenate all the offer texts into a single string separated by " | "
        offer_string = ' | '.join(offer_texts)
        
        # Append the offer string to the offers list
        offers.append(offer_string)
        
        # If no offer information is found, append 'N/A'
        if not offers:
            offers.append('N/A')

##warranty
        ## Warranty
        warranty_div = product.find('div', class_='_352bdz')
        if warranty_div:
            warranty_text = warranty_div.text.strip()
        else:
            warranty_text = 'N/A'
        warranty.append(warranty_text)
##variation
# Extract variations for the current product
        current_variations = {}
        for category in product.find_all('div', class_="_22QfJJ"):
            category_name = category.find('span').text
            options = [option.find('div', class_="_3Oikkn").text.strip() for option in category.find_all('li', class_="_3V2wfe")]
            if options:  # Check if there are any options found
                current_variations[category_name] = ', '.join(options)
            else:
                # If no options are found for this category, add a placeholder
                current_variations[category_name] = "N/A"
        
        # Check if current_variations dictionary is not empty before appending
        if current_variations:
            variations.append(current_variations)
        else:
            # If no variations were found for the product, append a placeholder
            variations.append({"N/A"})



##Highlights
        highlights_div = product.find('div', class_='_2418kt')
        if highlights_div:
            ul_element = highlights_div.find('ul')
            if ul_element:
                highlights_text = ul_element.get_text(separator=', ').strip()
                highlights.append(highlights_text)
        else:
            highlights.append('N/A')

##payment mode
        # Placeholder for payment modes
        payment_mode_value = 'N/A'  # Default value
        payment_modes_div = product.find('div', class_='_250Jnj')
        if payment_modes_div:
            payment_modes_list = payment_modes_div.find_all('li', class_='_1Ma4bX')
            if payment_modes_list:
                payment_mode_value = ', '.join([mode.text.strip() for mode in payment_modes_list])
        payment_modes.append(payment_mode_value)

##discription
# Initialize list outside the loop if not already done
        
        description_div = product.find('div', class_='_1mXcCf RmoJUa')
        
        if description_div:
            description_text = description_div.get_text(separator=' ', strip=True)
        else:
            description_text = 'N/A'  # Use 'N/A' if the specific div is not found
        
        # Append the description text for the current URL to the list
        description_list.append(description_text)
        

##A plus content       
        # Initialize variables to store concatenated text and image URLs
        content_dict = {
            "headings": [],
            "contents": [],
            "image_urls": []
            }
            
        # Find all divs that seem to represent individual items or sections
        for div in product.find_all("div", class_="K4SXrT funtru"):
            # Extract heading
            heading = div.find("div", class_="_3qWObK").text if div.find("div", class_="_3qWObK") else "N/A"
            content_dict["headings"].append(heading)
        
            # Extract paragraph text
            paragraph = div.find("div", class_="_3zQntF").text.strip() if div.find("div", class_="_3zQntF") else "N/A"
            content_dict["contents"].append(paragraph)
        
            # Extract image URL
            image_url = div.find("img")["src"] if div.find("img") else "N/A"
            content_dict["image_urls"].append(image_url)
            
        # Ensure lists are of consistent size by filling with 'N/A' as necessary
        max_length = max(len(content_dict["headings"]), len(content_dict["contents"]), len(content_dict["image_urls"]))
        for key in content_dict:
            while len(content_dict[key]) < max_length:
                content_dict[key].append("N/A")
            
        # Append the dictionary for the current URL to A_plus_contents
        A_plus_contents.append(content_dict)

##product specification
        # Initialize a temporary list for the current product's specifications
        temp_product_specification = []
        
        tables = product.find_all('table', class_='_14cfVK')
        for table in tables:
            rows = table.find_all('tr', class_='_1s_Smc row')
            for row in rows:
                row_data = []  # List to store the cells of the current row
                cells = row.find_all(['td', 'th'])
                for cell in cells:
                    cell_text = cell.text.strip()
                    row_data.append(cell_text)  # Append cell content to the row data list
                    
                temp_product_specification.append(row_data)  # Append row data to the temporary list
            
        # Append the temporary list as a single item to the main list
        product_specifications.append(temp_product_specification)


#Avg rating
        # Find the div element with class "_3LWZlK"
        div_element = product.find('div', class_='_3LWZlK')
            
        # Extract the text from the div element
        if div_element:
            avg_rating_text = div_element.get_text(strip=True)
            ratings_list.append(avg_rating_text)
        else:
            ratings_list.append('na')

##reviews and rating count
        # Use temporary placeholders for the current URL's data
        current_ratings = 'N/A'  # Placeholder if no data is found
        current_reviews = 'N/A'  # Placeholder if no data is found
        
        main_div = product.find('div', class_='gUuXy- _16VRIQ')
        if main_div:
            ratings_reviews_span = main_div.find('span', class_='_2_R_DZ')
            if ratings_reviews_span:
                ratings_reviews_text = ratings_reviews_span.get_text(separator=' ').strip()
                parts = ratings_reviews_text.split('&')
                if len(parts) == 2:
                    ratings_text, reviews_text = parts
                    
                    # Clean and extract the numerical part
                    ratings_count = ratings_text.strip().replace('Ratings', '').replace(',', '').strip()
                    reviews_count = reviews_text.strip().replace('Reviews', '').replace(',', '').strip()
                    
                    # Update the temporary placeholders with actual data
                    current_ratings = int(ratings_count)
                    current_reviews = int(reviews_count)
        
        # Append the data or placeholder for the current URL to the main lists
        count_of_ratings.append(current_ratings)
        count_of_reviews.append(current_reviews)

##Placeholder images
        # Find the div with class _2FHWw4
        div_container = product.find('div', class_='_2FHWw4')
        
        # Initialize an empty string to accumulate trimmed image URLs
        combined_urls = ""
        
        if div_container:
            # Find all img tags within the div container and trim their src attributes
            for img in div_container.find_all('img'):
                # Get the src attribute of the img tag
                original_url = img['src']
                
                # Replace "/128/128" with an empty string to remove it
                trimmed_url = original_url.replace("/128/128", "")
                
                # Append the trimmed URL to the combined_urls string, separated by a comma and a space
                combined_urls += trimmed_url + ", "
            
            # Remove the last comma and space to avoid an empty entry at the end
            combined_urls = combined_urls.rstrip(", ")
        
        else:
            # If div_container is None, just use 'N/A'
            combined_urls = 'N/A'
        
        # Store the combined URLs string or 'N/A' as the only element in the list
        placeholder_images.append(combined_urls)



        
##ratings by star
        star_ratings = []
        ratings_counts = []

        # Extracting star ratings
        star_elements = product.find_all('span', class_='_26f_zl')
        for star in star_elements:
            star_ratings.append(star.get_text())

        # Extracting ratings counts
        count_elements = product.find_all('div', class_='_1uJVNT')
        for count in count_elements:
            ratings_counts.append(count.get_text().replace(',', ''))

        # Combining star ratings and counts into a single list
        ratings_by_star = [(star, count) if star and count else ('N/A', 'N/A') for star, count in zip(star_ratings, ratings_counts)]

        # Create a list to store the ratings by star data for the current URL
        ratings_by_star_list = []
        for star, count in ratings_by_star:
            ratings_by_star_list.append({"Rating": star, "Count": count})

        # Append the ratings by star data to the main list
        all_ratings_by_star.append(ratings_by_star_list)

##reviews url
        #Find and extract the URL of the page with all the reviews
        try:
            # Find the reviews link element
            reviews_link = driver.find_element(By.XPATH, "//a[contains(@href, '/product-reviews/')]")


            
            # Get the reviews URL from the link
            reviews_url = reviews_link.get_attribute('href')
            
            # Append the reviews URL to the list
            reviews_urlss.append(reviews_url)
        except NoSuchElementException:
            # If reviews link is not found, append 'N/A'
            reviews_urlss.append('N/A')

##QnAs url
        #Find and extract the QnA of the page with all the QnA
        try:
            # Find the reviews link element
            QnAs_link = driver.find_element(By.XPATH, "//a[contains(@href, '/product-questions/')]")
            
            # Get the reviews URL from the link
            QnAs_urls = QnAs_link.get_attribute('href')
            
            # Append the reviews URL to the list
            QnAs_urlss.append(QnAs_urls)
        except NoSuchElementException:
            # If QnAs link is not found, append 'N/A'
            QnAs_urlss.append('N/A')


    else:
        # If the product is not found, append 'N/A' or appropriate placeholders to all lists
        #pids.append('N/A')
        product_titles.append('N/A')
        landing_image.append('N/A')
        category_node.append('N/A')
        fulfillment_mode.append('N/A')
        prices.append('N/A')
        MRPs.append('N/A')
        offers.append('N/A')
        warranty.append('N/A')
        variations.append('N/A')
        highlights.append('N/A')
        payment_modes.append('N/A')
        description_list.append('N/A')
        A_plus_contents.append('N/A')
        product_specifications.append({'Info': 'No specifications available'})
        ratings_list.append('No Rating')
        # Since count_of_ratings and count_of_reviews are initialized with ['N/A'], no need to append
        all_ratings_by_star.append([])
        reviews_urlss.append('N/A')
        QnAs_urlss.append('N/A')
        placeholder_images.append('N/A')
        count_of_ratings.append('N/A')
        count_of_reviews.append('N/A')
        stocks.append('In Stock')


    
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def html_code(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    return soup

def cus_rev(soup, pid):
    reviews = []
    review_blocks = soup.find_all('div', {'class': 'col _2wzgFH'})
    for block in review_blocks:
        rating_elem = block.find('div', {'class': '_3LWZlK'})
        review_elem = block.find('p', {'class': '_2-N8zT'})
        sum_elem = block.find('div', {'class': 't-ZTKy'})
        name_elem = block.find_all('p', {'class': '_2sc7ZR'})[0]
        date_elem = block.find_all('p', {'class': '_2sc7ZR'})[1]
        location_elem = block.find('p', {'class': '_2mcZGG'})
        
        if rating_elem and review_elem and name_elem and date_elem and sum_elem:
            review_description = sum_elem.text.strip().replace("READ MORE", "")
            
            review = {
                'PID': pid,
                'Name': name_elem.text.strip(),
                'Date': date_elem.text.strip(),
                'Rating': rating_elem.text,
                'Review': review_elem.text,
                'Review Description': review_description,
                'Location': location_elem.text if location_elem else 'N/A',
            }
            reviews.append(review)
    return reviews

def scrape_reviews(product_urls):
    all_reviews = []

    for url in tqdm(product_urls):
        # Extract PID from URL
        pid = url.split('pid=')[1].split('&')[0]
        
        print("Scraping URL:", url)
        soup = html_code(url)
        page_reviews = cus_rev(soup, pid)  # Pass PID to cus_rev function
        all_reviews.extend(page_reviews)

        # Introduce a delay
        time.sleep(1)

    return all_reviews

# Example usage

all_reviews = scrape_reviews(product_urls)

# Convert reviews to DataFrame
df_reviews1 = pd.DataFrame(all_reviews)

# Print or save the DataFrame as needed
print(df_reviews1)
```

```python
import requests
from bs4 import BeautifulSoup
from datetime import date
import pandas as pd

def html_code(url):
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'}
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, "html.parser")
    return soup

def cus_rev(soup, pid):
    reviews = []
    review_blocks = soup.find_all('div', {'class': '_27M-vq'})
    for block in review_blocks:
        rating_elem = block.find('div', {'class': '_3LWZlK'})
        review_elem = block.find('p', {'class': '_2-N8zT'})
        sum_elem = block.find('div', {'class': 't-ZTKy'})
        name_elem = block.find_all('p', {'class': '_2sc7ZR'})[0]
        date_elem = block.find_all('p', {'class': '_2sc7ZR'})[1]
        location_elem = block.find('p', {'class': '_2mcZGG'})
        
        if rating_elem and review_elem and name_elem and date_elem and sum_elem:
            # Remove "READ MORE" from the review description
            review_description = sum_elem.text.strip().replace("READ MORE", "")
            
            review = {
                'PID': pid,
                'Name': name_elem.text.strip(),
                'Date': date_elem.text.strip(),
                'Rating': rating_elem.text,
                'Review': review_elem.text,
                'Review Description': review_description,
                'Location': location_elem.text if location_elem else 'N/A',  # Added check for location
            }
            reviews.append(review)
    return reviews


def scrape_reviews(review_urls):
    all_reviews = []

    for url in review_urls:
        # Extract PID from URL
        pid = url.split('pid=')[1].split('&')[0]
        
        page = 1
        while True:
            page_url = url + "&page=" + str(page)
            print(page_url)
            print(page, "st page is scraping")
            soup = html_code(page_url)
            page_reviews = cus_rev(soup, pid)  # Pass PID to cus_rev function
            if not page_reviews:
                break
            all_reviews.extend(page_reviews)
            page += 1

            # Introduce a delay
            time.sleep(2)

    return all_reviews

# Scrape reviews from multiple URLs
all_reviews = scrape_reviews(reviews_urls)

# Convert reviews to DataFrame
df_reviews2 = pd.DataFrame(all_reviews)

# Print the DataFrame
print(df_reviews2)
    
```
